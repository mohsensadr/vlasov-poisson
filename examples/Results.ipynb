# Simulation Results

```python
import numpy as np
import matplotlib.pyplot as plt

t = "050"

fig, axes = plt.subplots(3, 2, figsize=(8, 12))  # 3 rows, 2 columns

path = "/content/vlasov-poisson/bin/data/"

x = np.loadtxt(path+"N_step_"+t+".csv", delimiter=",")
im = axes[0,0].imshow(x, origin="lower", cmap="plasma")
fig.colorbar(im, ax=axes[0, 0])
axes[0,0].set_ylabel(r"$y\ [-]$")
axes[0, 0].set_title(r"$\langle N \rangle$")

x = np.loadtxt(path+"NVR_step_"+t+".csv", delimiter=",")
axes[0,1].imshow(x, origin="lower", cmap="plasma")
im = axes[0, 1].imshow(x, origin="lower", cmap="plasma")
fig.colorbar(im, ax=axes[0, 1])
axes[0, 1].set_title(r"$\langle N \rangle_\mathrm{VR}$")

###################

x = np.loadtxt(path+"Ux_step_"+t+".csv", delimiter=",")
im = axes[1,0].imshow(x, origin="lower", cmap="plasma")
fig.colorbar(im, ax=axes[1, 0])
axes[1,0].set_ylabel(r"$y\ [-]$")
axes[1, 0].set_title(r"$\langle U_x \rangle$")

x = np.loadtxt(path+"UxVR_step_"+t+".csv", delimiter=",")
axes[1,1].imshow(x, origin="lower", cmap="plasma")
im = axes[1, 1].imshow(x, origin="lower", cmap="plasma")
fig.colorbar(im, ax=axes[1, 1])
axes[1, 1].set_title(r"$\langle U_x \rangle_\mathrm{VR}$")

###################

x = np.loadtxt(path+"T_step_"+t+".csv", delimiter=",")
im = axes[2,0].imshow(x, origin="lower", cmap="plasma")
fig.colorbar(im, ax=axes[2, 0])
axes[2,0].set_ylabel(r"$y\ [-]$")
axes[2,0].set_xlabel(r"$x\ [-]$")
axes[2, 0].set_title(r"$\langle T \rangle$")

x = np.loadtxt(path+"TVR_step_"+t+".csv", delimiter=",")
axes[2,1].imshow(x, origin="lower", cmap="plasma")
im = axes[2, 1].imshow(x, origin="lower", cmap="plasma")
fig.colorbar(im, ax=axes[2, 1])
axes[2,1].set_xlabel(r"$x\ [-]$")
axes[2, 1].set_title(r"$\langle T \rangle_\mathrm{VR}$")

#
```
